### Setting Styling in Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
# plotting multiple sub plots
fig, ax = plt.subplots(3)  # 3 tells us on how many sub plots we would be needing

# ax is the array of length 3 where each is a 2axis object
ax[0].plot(x, np.sin(x));
ax[1].plot(x, np.cos(x));
ax[2].plot(x, np.tan(x));
```

### Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
# always the first steps
fig = plt.figure()
ax = plt.axes()

# all the plots that you make will be part of axis
x = np.linspace(0, 20, 100)
ax.plot(x, np.sin(x), color='red', linestyle='dashed');
ax.plot(x, np.cos(x), color='green', linestyle='dashed');
ax.set_xlim(0, 20);
ax.set_ylim(-1.5, 1.5);
ax.set_title('Sin(x) & Cos(x)');
ax.set_xlabel('x');
ax.set_ylabel('range');

# linestyle takes (dashed, dotted, solid, dashdot)
# color: takes care of what chart has to be made

#xlim, ylim is used to set min/max range in x and y axis respectively

#title, xlabel, ylabel are used to set title, x label and y label of the plot in making.
```

### Scatter Chart

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fast')
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.3,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

iris = load_iris()
feat = scaler.fit_transform(iris.data)
features = feat.T

plt.scatter(features[0], features[1], alpha=0.3,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
